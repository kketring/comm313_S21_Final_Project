# Sentiment Analysis – What Does Twitter Think?

In this next section, we will be using AI to guage whether sentiment on Twitter has indeed become more negative and  will do this by using Sentiment Analysis. See below for a full description of Sentiment Analysis. 

> * __Sentiment Analysis (SA)__ is a broad term used to describe NLP approaches to quantifying the types and degree of emotions expressed in text.
>
> * There are two broad approaches to SA that we will consider:
>
>     1. __LEXICON-BASED SA__ - A list of words (_lexicon_) are scored on various dimensions (e.g. valence, arousal) or emotions (e.g. happiness, anger, sadness). A compound score is calculated based on the proportion of words in the target text that belong to each category or based on the mean score of all the words in the target text that appear in the lexicon.
    2. __CLASSIFIER SA__ - Machine learning algorithm is used to create a pattern of features that are associated with pre-labeled texts. For example, a sample of positive and negative reviews (based on 5-star rating for instance).

This analysis will be utilizing the Lexicon-Based SA and specifically looking at the compound score. The most positive compound score is +1 and the most negative compound score is -1.
